```python
sci_dict = {'Video ID':[], 'Title':[],'Description':[], 'Category':[]}
```

```python
ctr = 0
```

```python
def grabVideos(keyword,token=None):
    res = youtube_search(keyword)
    token = res[0]
    videos = res[1]
    for vid in videos:
        sci_dict['Video ID'].append(vid['id']['videoId'])
        sci_dict['Title'].append(vid['snippet']['title'])
        sci_dict['Description'].append(vid['snippet']['description'])
        sci_dict['Category'].append('technology')
    print("added " + str(len(videos)) + " videos to a total of " + str(len(sci_dict['Video ID'])))
    return token
```

```python
token = grabVideos("science and technology")
while ctr<=36:
    token = grabVideos("science and technology", token=token)
    ctr+=1
```

```python
sci_dict
```

```python
import csv
```

```python
with open("technology.csv", "w") as outfile:
    writer = csv.writer(outfile)
    writer.writerow(sci_dict.keys())
    writer.writerows(zip(*sci_dict.values()))
```
